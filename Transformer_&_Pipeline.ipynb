```python
from transformers import pipeline

classifier=pipeline("sentiment-analysis")
classifier("This is the best day in my life") # can use array of words in classifier with array
```

```python
from transformers import pipeline
classifier =pipeline("zero-shot-classification")  # for categorical data only
classifier("I am computerscience student",
           candidate_labels=['education','business','politics'])
```

```python
generator=pipeline('text-generation')  # without any model
generator("I am Pooja.I curruently pursuing")
```

```python
from transformers import pipeline
generator=pipeline('text-generation', model="K-intelligence/Midm-2.0-Mini-Instruct")  # with any model
generator("I am Pooja.I curruently pursuing")
```

```python
from transformers import pipeline

unmasker =pipeline('fill-mask')  # used to fill words
unmasker("This is life with full of <mask>.",top_k=2) # top_k is to fill words from top
```

```python
ner=pipeline('ner',grouped_entities=True)  # ner --> name , entity, recognization --> used to fetch name, organization and place
ner("My name is pooja, studying CSE at karpagam University at coimbatore")
```

```python
ques_ans=pipeline("question-answering")
ques_ans(
    question="when did Gandhi born?",
    context="""Mahatma Gandhi, born Mohandas Karamchand Gandhi on
    October 2, 1869, in Porbandar, India, was one of the most influential leaders in the world’s history.
    Revered as the Father of the Nation in India, Gandhi led a nonviolent struggle against British colonial rule,
    shaping the Indian independence movement with his unique philosophy of truth (Satya) and nonviolence (Ahimsa).""",
)
```

```python
summarizer = pipeline("summarization")
summarizer(
    """
    Sure, here's a concise and meaningful piece about **Mahatma Gandhi** — suitable for speeches, essays, or general knowledge:

---

**Mahatma Gandhi: The Father of the Nation**

Mahatma Gandhi, born Mohandas Karamchand Gandhi on October 2, 1869, in Porbandar, India, was one of the most influential leaders in the world’s history. Revered as the **Father of the Nation** in India, Gandhi led a nonviolent struggle against British colonial rule, shaping the Indian independence movement with his unique philosophy of **truth (Satya)** and **nonviolence (Ahimsa)**.

Educated as a lawyer in London, Gandhi began his activism in South Africa, where he fought against racial discrimination. Upon returning to India, he mobilized millions through peaceful protests, civil disobedience, and the famous **Salt March** of 1930. His movements, like **Non-Cooperation** and **Quit India**, challenged British authority without lifting a weapon.

What made Gandhi remarkable was not just his political vision but his deep commitment to simple living, self-discipline, and equality for all — regardless of religion, caste, or gender. His principles continue to inspire global movements for civil rights and social justice.

On January 30, 1948, Gandhi was assassinated, but his legacy lives on as a symbol of peace, moral courage, and the power of nonviolence.

---

Would you like this turned into a poem, children's version, or more detailed historical piece?

    """
)
```
